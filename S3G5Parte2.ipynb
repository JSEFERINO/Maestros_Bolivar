```python
# @title **Volumen y area de un cono** $$V_{cono}(r,h)=\frac{1}{3}\pi*r^2*h$$  $$A_{cono}(r,h)=\pi*r*(r+\sqrt{r^2+h^2})$$

# Parmetros
r = float(input('Ingrese el Radio del cono r = '))    # radio
h = float(input('Ingrese la Altura del cono h = '))    # altura

# Volumen del cono
vol_cono = (1/3) * np.pi * r**2 * h

# rea superficial del cono
area_cono = np.pi * r * (r + np.sqrt(r**2 + h**2))

# Resultados
print(f"El volumen del cono es: V({r},{h}) = {vol_cono}" )
print(f"El rea superficial del cono es: A({r},{h}) = {area_cono}" )

```

```python
# prompt: haz la grafica de la esfera del paso anterior

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def sphere_surface(r):
  """Generates points on the surface of a sphere with radius r."""
  u = np.linspace(0, 2 * np.pi, 100)
  v = np.linspace(0, np.pi, 100)
  x = r * np.outer(np.cos(u), np.sin(v))
  y = r * np.outer(np.sin(u), np.sin(v))
  z = r * np.outer(np.ones(np.size(u)), np.cos(v))
  return x, y, z

# Example usage:
radius = 5.385164807134504  # You can change the radius here

x, y, z = sphere_surface(radius)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Sphere with Radius {}'.format(radius))

plt.show()
```

# **e. Ejercicio: Halle todas las raices de los polinomios:**

* $P(x)= x^2-x-6$
* $P(x) = x^3 + 3x^2-4x-12$
* $P(x) = x^4-10x^2 + 9$

```python
# @title **Teorema Fundamental de la Aritmtica**

def descomponer_en_primos(n):
  """Descompone un nmero entero en producto de potencias de primos.

  Args:
    n: El nmero entero a descomponer.

  Returns:
    Una lista de tuplas (primo, exponente) que representa la descomposicin
    en primos.
  """
  factores = []
  i = 2
  while i * i <= n:
    if n % i:
      i += 1
    else:
      n //= i
      exponente = 1
      while n % i == 0:
        n //= i
        exponente += 1
      factores.append((i, exponente))
  if n > 1:
    factores.append((n, 1))
  return factores

# Solicitar al usuario que ingrese un nmero
numero = int(input("Ingrese un nmero entero: "))

# Descomponer el nmero en producto de potencias de primos
descomposicion = descomponer_en_primos(numero)

# Mostrar la descomposicin
if not descomposicion:
  print(f"{numero} es un nmero primo.")
else:
  print(f"La descomposicin en factores primos de {numero} =:")
  for primo, exponente in descomposicion:
    if exponente == 1:
      print(f"{primo}", end="")
    else:
      print(f"{primo}^{exponente}", end="")
    if primo != descomposicion[-1][0]:
      print(" * ", end="")
  print()
```

$\bullet$ **Paso_10. Todo el algoritmo en un solo lugar pero con las letras que aparecen en el prrafo**

```python
import matplotlib.pyplot as plt
import pandas as pd
import string

print("Letra   Frecuencia")
print("-----------------")
# Prrafo
parrafo = 'La ciencia de datos es un campo acadmico interdisciplinario que utiliza estadstica, computacin cientfica, mtodos, procesos, algoritmos y sistemas cientficos para obtener (recolectar o extraer), tratar, analizar y presentar informes a partir de datos ruidosos, estructurados y no estructurados.'

# Convertir el prrafo a minsculas para contar letras sin distincin entre maysculas y minsculas
parrafo = parrafo.lower()

# Crear un diccionario para almacenar la frecuencia de cada letra
frecuencia_letras = {}

# Iterar a travs del prrafo y contar las letras
for letra in parrafo:
    if letra.isalpha():  # Verificar si es una letra
        if letra in frecuencia_letras:
            frecuencia_letras[letra] += 1
        else:
            frecuencia_letras[letra] = 1

# Mostrar la tabla de frecuencia de letras

for letra, frecuencia in frecuencia_letras.items():
    print(f"{letra}      {frecuencia}")


# Crear un DataFrame a partir del diccionario de frecuencias
df = pd.DataFrame(frecuencia_letras.values(), index=frecuencia_letras.keys(), columns=['Frecuencia'])

# Crear un grfico de barras
plt.figure(figsize=(10, 4))
df.plot(kind='bar', legend=False)
plt.xlabel('Letra')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de letras en el prrafo')
plt.xticks(rotation=0)
plt.show()

# Crear un grfico de pastel
plt.figure(figsize=(6, 6))
df.plot(kind='pie', y='Frecuencia', autopct='%1.1f%%', legend=False)
plt.title('Frecuencia de letras en el prrafo')
plt.show()
```
