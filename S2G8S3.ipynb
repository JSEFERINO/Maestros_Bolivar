# **8. Volumen y Areas de sólidos**

```python
# @title **a. Volumen de la Esfera** $$V(r)=\frac{4}{3}\pi*r^3$$
# Parametro de Esfera
r = float(input("Ingrese el radio de la esfera r = "))

# Volumen de la Esfera
Vol_esfera = (4/3)*np.pi*r**3

# Imprimiendo el volumen de la esfera
print(f'El volumen de la esfera es V({r}) = {Vol_esfera:.2f}')
```

```python
# @title **b. Volumen y area de un cono** $$V_{cono}(r,h)=\frac{1}{3}\pi*r^2*h$$  $$A_{cono}(r,h)=\pi*r*(r+\sqrt{r^2+h^2})$$
# Parametros de un cono
r = float(input("Ingrese el radio de la esfera r = "))
h = float(input("Ingrese la altura de la esfera h = "))

# Volumen de un cono
Vol_cono = (1/3)*np.pi*r**2*h

# Imprimiendo el volumen de un cono
print(f'El volumen del cono es V({r},{h}) = {Vol_cono:.2f}')

# Area del cono
Area_cono = np.pi*r*(r+np.sqrt(r**2+h**2))

# Imprimiendo el volumen de un cono
print(f'El área del cono es A({r},{h}) = {Area_cono:.2f}')
```

```python
# @title **c. Hallar el volumen y area del cono de radio r = 3 y altura h = 5**
r = 3
h = 5
# Volumen de un cono
Vol_cono = (1/3)*np.pi*r**2*h

# Imprimiendo el volumen de un cono
print(f'El volumen del cono es V({r},{h}) = {Vol_cono:.2f}')

# Area del cono
Area_cono = np.pi*r*(r+np.sqrt(r**2+h**2))

# Imprimiendo el volumen de un cono
print(f'El área del cono es A({r},{h}) = {Area_cono:.2f}')



```

```python

# @title **d. Volumen y area de un cilindro** $$V_{cilindro}(r,h)=\pi*r^2*h$$  $$A_{cilindro}(r,h)=2*\pi*r*(r+h)$$


# Parametros de un cilindro
r = float(input("Ingrese el radio del cilindro r = "))
h = float(input("Ingrese la altura del cilindro h = "))

# Volumen del cilindro
Vol_cilindro = np.pi*r**2*h

# Imprimiendo el volumen del cilindro
print(f'El volumen del cilindro es V({r},{h}) = {Vol_cilindro:.2f}')

# Area del Cilindro
Area_cilindro = 2*np.pi*r*(r+h)

# Imprimiendo Area del Cilindro
print(f'El área del cilindro es A({r},{h}) = {Area_cilindro:.2f}')
```

```python
# @title **b. Compuesta de f y g en 2**

def f(x):
  return x**2 + 3*x + 1

def g(x):
  return x**3 - 2*x**2 + 5*x - 7

def fog(x):
  return f(g(x))

print(f"La compuesta de f y g en 2 es: {fog(2)}")
```

```python
# @title **c. Graficar $$f(x)=x^2 + 3x + 1$$**

import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return x**2 + 3*x + 1

x = np.linspace(-10, 10, 100)  # Crea un rango de valores de x
y = f(x)  # Calcula los valores correspondientes de y

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfica de f(x) = x^2 + 3x + 1')
plt.grid(True)
plt.show()
```

```python
# @title **d. Graficar $$g(x)=x^3 - 2x^2 + 5x - 7$$**

import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return x**3 -2*x**2+5*x -7

x = np.linspace(-10, 10, 100)  # Crea un rango de valores de x
y = f(x)  # Calcula los valores correspondientes de y

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('g(x)')
plt.title('Gráfica de g(x)= x^3 - 2x^2 + 5x - 7')
plt.grid(True)
plt.show()
```

```python
# @title **e. Graficar f(x)+g(x)**

def f(x):
  return x**2 + 3*x + 1

def g(x):
  return x**3 - 2*x**2 + 5*x - 7

def f_plus_g(x):
  return f(x) + g(x)

x = np.linspace(-10, 10, 100)  # Crea un rango de valores de x
y = f_plus_g(x)  # Calcula los valores correspondientes de y

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x) + g(x)')
plt.title('Gráfica de f(x) + g(x)')
plt.grid(True)
plt.show()
```

```python
# Calcular el discriminante
D = b**2 - 4*a*c
print(f'El discriminante del polinomio es D =({b})^2-4*({a})*({c}) = {D}')
```

$\bullet$ **Paso_9. Todo el algoritmo en un solo lugar**

```python
import matplotlib.pyplot as plt
import pandas as pd
import string


# Convertir el párrafo a minúsculas para evitar distinción entre mayúsculas y minúsculas
parrafo = parrafo.lower()

# Crear una lista de letras en el alfabeto
alfabeto = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']

# Inicializar un diccionario para almacenar la frecuencia de cada letra
frecuencias = {letra: 0 for letra in alfabeto}

# Contar la frecuencia de cada letra en el párrafo
for letra in parrafo:
    if letra in alfabeto:
        frecuencias[letra] += 1

# Crear un DataFrame a partir del diccionario de frecuencias
df = pd.DataFrame(frecuencias.values(), index=frecuencias.keys(), columns=['Frecuencia'])

# Crear un gráfico de barras
plt.figure(figsize=(10, 4))
df.plot(kind='bar', legend=False)
plt.xlabel('Letra')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de letras en el párrafo')
plt.xticks(rotation=0)
plt.show()

# Crear un gráfico de pastel
plt.figure(figsize=(6, 6))
df.plot(kind='pie', y='Frecuencia', autopct='%1.1f%%', legend=False)
plt.title('Frecuencia de letras en el párrafo')
plt.show()
```

$\bullet$ **Paso_10. Todo el algoritmo en un solo lugar pero con las letras que aparecen en el párrafo**

```python
import matplotlib.pyplot as plt
import pandas as pd
import string

print("Letra   Frecuencia")
print("-----------------")
# Párrafo
parrafo = 'La ciencia de datos es un campo académico interdisciplinario que utiliza estadística, computación científica, métodos, procesos, algoritmos y sistemas científicos para obtener (recolectar o extraer), tratar, analizar y presentar informes a partir de datos ruidosos, estructurados y no estructurados.'

# Convertir el párrafo a minúsculas para contar letras sin distinción entre mayúsculas y minúsculas
parrafo = parrafo.lower()

# Crear un diccionario para almacenar la frecuencia de cada letra
frecuencia_letras = {}

# Iterar a través del párrafo y contar las letras
for letra in parrafo:
    if letra.isalpha():  # Verificar si es una letra
        if letra in frecuencia_letras:
            frecuencia_letras[letra] += 1
        else:
            frecuencia_letras[letra] = 1

# Mostrar la tabla de frecuencia de letras

for letra, frecuencia in frecuencia_letras.items():
    print(f"{letra}      {frecuencia}")


# Crear un DataFrame a partir del diccionario de frecuencias
df = pd.DataFrame(frecuencia_letras.values(), index=frecuencia_letras.keys(), columns=['Frecuencia'])

# Crear un gráfico de barras
plt.figure(figsize=(10, 4))
df.plot(kind='bar', legend=False)
plt.xlabel('Letra')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de letras en el párrafo')
plt.xticks(rotation=0)
plt.show()

# Crear un gráfico de pastel
plt.figure(figsize=(6, 6))
df.plot(kind='pie', y='Frecuencia', autopct='%1.1f%%', legend=False)
plt.title('Frecuencia de letras en el párrafo')
plt.show()
```
